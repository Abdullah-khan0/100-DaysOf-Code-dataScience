# Matplotlib Markers

Mark each point with a circle:

Mark each point with a Star:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, marker = '*')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, 'o:b') # o = circle, b = blue, : = dotted line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, marker = 'h', color='g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, marker = 'p', color='r')
plt.show()
```

# Marker Size,Color

There are two parameters available to colorize the matplotlib markers:

Matplotlib Marker Edge Color (mec).

Matplotlib Marker Face Color (mfc).

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, marker = '*', ms = 25)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, 'o:r', ms = 20,mec='b' )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, 'o:r', ms = 20,mec='b',mfc='b' )
plt.show()
```
