# Matplotlib Scatter

Creating Scatter Plots

Color Each Dot

ColorMap

Size

Alpha

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 7])
plt.scatter(x, y)

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.show()
```

```python
#color each dot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 7])
colors = np.array(["red","green","blue","pink","gray"])
plt.scatter(x, y, c = colors)

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.show()
```

```python
#colormap
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 7])
colors = np.array([0, 10,20,30,40])
plt.scatter(x, y, c = colors, cmap='viridis')

plt.colorbar()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.show()
```

```python
#size, alpha
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 7])
sizes = np.array([20,50,100,200,500])
plt.scatter(x, y, alpha = 0.8, s=sizes)

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.show()
```

- Combine Color Size and Alpha

Create random arrays with 100 values for x-points, y-points, colors and sizes:


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='spring')

plt.colorbar()

plt.show()
```

# Matplotlib Bars Plots

Creating Bars

Horizontal Bars

Bar Color

Bar Width

Bar Height


```python
#Creating Bars

import matplotlib.pyplot as plt
import numpy as np
x = np.array([' A', ' B', ' C', ' D'])
y = np.array([25, 40, 30, 50])

plt.bar(x, y)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title(' Bar Plot')
plt.show()
```

```python
#Horizontal Bars

import matplotlib.pyplot as plt
import numpy as np
x = np.array([' A', ' B', ' C', ' D'])
y = np.array([25, 40, 30, 50])

plt.barh(x, y, color="hotpink")#for horizontal write plt.barh()

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title(' Bar Plot')
plt.show()
```

Bar Width

The bar() takes the keyword argument width to set the width of the bars:

For horizontal bars, use height instead of width.

```python
#Creating Bars

import matplotlib.pyplot as plt
import numpy as np
x = np.array([' A', ' B', ' C', ' D'])
y = np.array([25, 40, 30, 50])

plt.bar(x, y, width=0.2)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title(' Bar Plot')
plt.show()
```

# Matplotlib Histograms


A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

Example: Say you ask for the height of 250 people, you might end up with a histogram 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

# Matplotlib Pie Charts


Creating Pie Charts

Labels

Start Angle

Explode

Shadow

Colors

Legend

Legend With Header

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = fruits, startangle = 90)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]
explod = [0.2, 0, 0, 0] #Explode

plt.pie(y, labels = fruits, explode = explod , shadow=True)#shadow
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]
color = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = fruits, colors = color)
plt.legend(title="Four Fruits")
plt.show() 
```
