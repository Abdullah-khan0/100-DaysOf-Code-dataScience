# Colour images
In colour image each pixel is represented by 3 values - RGB

##### [0,0,0]: Black

##### [255, 255,255]: White

##### [255,0,0]: Red

##### [0,255,0]: Green

##### [0,0,255]: Blue

##### [100,0,0]: Dark Red 

```python
# Let us generate colour image of 4 x 2 pixels
y=np.array([[[255,0,0],[0,255,140]],
           [[0,255,0],[255,255,0]],
           [[0,0,255],[255,255,255]],
           [[0,0,0],[0,200,2]]])
plt.imshow(y)
y.shape 
```

## Image Processing

```python
img=plt.imread("/kaggle/input/msdhoni/ms-dhoni.jpg")
print(img.shape) # printing shape of image. It is colour so 427 x 630 image having 3 channel - RGB
plt.imshow(img)
plt.show()
```

## Crop Image

```python
# Crop image
img2=img[100:800,400:1000]
plt.imshow(img2)
plt.show()
```

## Increase Brightness

```python
# Increase the brightness
BrightnessUp = 100 * np.ones((img.shape),dtype='int32')
img3=img+BrightnessUp
plt.imshow(img3)
plt.show()
```

## Decrease the brightness

```python
# Decrease the brightness
BrightnessDown = -100 * np.ones((img.shape),dtype='int32')
img3=img+BrightnessDown
plt.imshow(img3)
plt.show()
```

## Add noise

```python
# Add noise to the image
AddNoise=np.random.randint(0,100,img.shape) # generating random number between 0 to 100 for the noise value
img4=img+AddNoise
plt.imshow(img4)
plt.show()
```

# Displaying image using OpenCV

```python
import cv2
```

## Reading image using OpenCv, this is default BGR format

```python
imgOCV=cv2.imread("/kaggle/input/msdhoni/ms-dhoni.jpg") 

plt.imshow(imgOCV) 
plt.show()
```

## convert image read by openCV in BRG format to RGB format

```python
imgOCV1=cv2.cvtColor(imgOCV,cv2.COLOR_BGR2RGBA)
plt.imshow(imgOCV1)
plt.show()
```

## Reading image in grayscale using openCV

```python
# Convert to grayscale
imgGray = cv2.cvtColor(imgOCV, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(imgGray, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

## Detecting boundries of the objects in image

```python
imgOCV3 = cv2.Canny(imgOCV,150,255)

plt.imshow(imgOCV3, cmap='gray') 
plt.show() 
```

## Image Bluring

```python
# Display the original image
plt.imshow(cv2.cvtColor(imgOCV, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Resize the image (shrink by a factor of 0.5 in both dimensions)
shrunken_img = cv2.resize(imgOCV, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_AREA)

# Display the shrunken image
plt.imshow(cv2.cvtColor(shrunken_img, cv2.COLOR_BGR2RGB))
plt.title('Shrunken Image')
plt.show()
```

```python
# Blurring - smoothes the image out
imgOCV8 = imgOCV1

blur = cv2.blur(imgOCV8,(18, 18))
gblur = cv2.GaussianBlur(imgOCV8,(7,7),0)

plt.imshow(imgOCV8), plt.title('Original Image')
plt.show()
plt.imshow(blur), plt.title('Blurred Image')
plt.show()
plt.imshow(gblur),plt.title('Gaussian Blurred Image')
plt.show()
```
